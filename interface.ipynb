```python
def fenetre():
    global window1
    window1=Tk()
    window1.title("Interface Graphique")
    frame=Frame(window1)
    frame.pack()
    #calculatrice
    
    form=LabelFrame(frame, text="Formulaire")
    form.grid(row=0, column=0, padx=20, pady=20)

    firstName=Label(form, text="Nom:")
    firstName.grid(row=0, column=0)
    
    value1 = StringVar() 
    firstName_entry=Entry(form,textvariable=value1)
    firstName_entry.grid(row=0, column=1)
    #firstName_entry.place(x=100, y=100)

    lastName=Label(form, text="PreNom:")
    lastName.grid(row=1, column=0)

    lastName_entry=Entry(form)
    lastName_entry.grid(row=1, column=1)

    country=Label(form, text="Pays:")
    #state empeche les utilisateur d'entrer leur propre valeur
    country_combo=Combobox(form, values=["Cameroun", "Nigeria","Tchad", "Congo"], state="readonly")
    #index de l"element qui s'affiche en premier
    country_combo.current(0)
    country.grid(row=2, column=0)
    country_combo.grid(row=2, column=1)
    #emet l'evenement lorsque la valeur selectionnée change
    country_combo.bind("<<ComboboxSelected>>")

    valeur_cochee1=StringVar()
    valeur_cochee2=StringVar()
    valeur_cochee3=StringVar()
    sport=Label(form, text="Sport préferé:")
    sport_check1=Checkbutton(form, text="Football",variable=valeur_cochee1, onvalue="football",offvalue="")
    sport_check2=Checkbutton(form, text="Handball", variable=valeur_cochee2, onvalue="Handball", offvalue="")
    sport_check3=Checkbutton(form, text="Basketball", variable=valeur_cochee3, onvalue="Basketball", offvalue="")
    #sort_check1.var = BooleanVar(value=True)
    sport.grid(row=3, column=0)
    sport_check1.grid(row=3, column=1)
    sport_check2.grid(row=3, column=2)
    sport_check3.grid(row=3, column=3) 


    sexe=Label(form, text="Sexe:")
    v = StringVar()
     # initialiser
    sexe_radio=Radiobutton(form, text="M", variable=v, value="masculin")
    sexe_radio1=Radiobutton(form, text="F",variable=v, value="feminin")
    sexe.grid(row=4, column=0)
    sexe_radio.grid(row=4, column=1)
    sexe_radio1.grid(row=4, column=2)
    sexe_radio.invoke()

    Password=Label(form, text="Password:")
    Password.grid(row=5, column=0)
    Password_entry=Entry(form)
    Password_entry.config(show = "*")
    Password_entry.grid(row=5, column=1)

    Annuler=Button(form, text="Annuler", command=reset)
    Annuler.grid(row=6, column=0)

    Envoyer=Button(form, text="Envoyer", command=send)
    Envoyer.grid(row=6, column=1)

    calcul=Button(form, text="calculatrice", command=calculator)
    calcul.grid(row=6, column=2)
    
    resultat=LabelFrame(frame, text="operateur")
    resultat.grid(row=7, column=0, sticky="news" , padx=20, pady=20)
    f_name1=Label(resultat)
    f_name1.grid(row=7, column=0)


    #resultat=Label(form)
    
    window1.mainloop()
```

```python
if __name__ == '__main__':
    fenetre()
```

```python
def addition():
    N1=float(nombre1_entry.get())
    N2=float(nombre2_entry.get())
    div=N1+N2
    resultat_entry.delete(0, END)
    resultat_entry.insert(0, div) 
```

```python
def soustraction():
    N1=float(nombre1_entry.get())
    N2=float(nombre2_entry.get())
    div=N1-N2
    resultat_entry.delete(0, END)
    resultat_entry.insert(0, div) 
    
```

```python
def multiplication():
    N1=float(nombre1_entry.get())
    N2=float(nombre2_entry.get())
    div=N1*N2
    resultat_entry.delete(0, END)
    resultat_entry.insert(0, div) 
    
```

```python
def division():
    N1=float(nombre1_entry.get())
    N2=float(nombre2_entry.get())
    if(N2==0):
        resultat_entry.delete(0, END)
        resultat_entry.insert(0, 'division impossible')
       
    else:
        div=N1/N2
        resultat_entry.delete(0, END)
        resultat_entry.insert(0, div)
        
```

```python
window=Tk()
window.title("Interface Graphique")
window.geometry('450x450')
#window.configure(bg='black') 
frame=Frame(window)
frame.pack()
#calculatrice
calculatrice=LabelFrame(frame, text="Calculatrice") 
calculatrice.grid(row=0, column=0)

bloc_nombre=LabelFrame(calculatrice)
bloc_nombre.grid(row=0, column=0)
nombre1=Label(bloc_nombre, text="Nombre 1:")
nombre1.grid(row=0, column=1)

nombre1_entry=Entry(bloc_nombre, width=30)
nombre1_entry.grid(row=0, column=2, padx=15)

nombre2=Label(bloc_nombre, text="Nombre 2:")
nombre2.grid(row=1, column=1)

nombre2_entry=Entry(bloc_nombre, width=30)
nombre2_entry.grid(row=1, column=2, padx=15, pady=20)

#operateur
operateur=LabelFrame(calculatrice, text="operateur")
operateur.grid(row=1, column=0, sticky="news" , padx=20, pady=20)

bloc_oper_add=LabelFrame(operateur)
bloc_oper_add.grid(row=1, column=1)
operateur1=Button(bloc_oper_add, width=9,text="+", command=addition)
operateur1.grid(row=0, column=0)

bloc_oper_sous=LabelFrame(operateur)
bloc_oper_sous.grid(row=1, column=2)
operateur2=Button(bloc_oper_sous, width=9,text="-", command=soustraction)
operateur2.grid(row=0, column=0)

bloc_oper_mult=LabelFrame(operateur)
bloc_oper_mult.grid(row=1, column=3)
operateur3=Button(bloc_oper_mult, width=9, text="*", command=multiplication)
operateur3.grid(row=0, column=0)

bloc_oper_div=LabelFrame(operateur)
bloc_oper_div.grid(row=1, column=4)
operateur4=Button(bloc_oper_div, width=9, text="/", command=division) 
operateur4.grid(row=0, column=0)
#operateur
#resultat=LabelFrame(calculatrice, text="resultat")
resultat_entry=Entry(calculatrice, width=30)
resultat_entry.grid(row=2, column=0,  padx=10, pady=20)

#result=Label(resultat, text="Nombre 2:")
#result.grid(resultat, row=2, column=0)
window.mainloop()
```
